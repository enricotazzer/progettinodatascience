```python
# ricerca del suo tiro più efficiente
shot_type = shot.groupby('ACTION_TYPE').sum().sort_values(by='PLAYER_NAME', ascending=False).loc[:, ['SHOT_ATTEMPTED_FLAG', 'SHOT_MADE_FLAG']]
shot_type['AVG'] = shot_type['SHOT_MADE_FLAG'] / shot_type['SHOT_ATTEMPTED_FLAG']
significativi = shot_type['SHOT_ATTEMPTED_FLAG'] > 200
shot_type[significativi].sort_values(by='SHOT_MADE_FLAG', ascending=False)

#raggruppiamo in tre grandi gruppi -> jump, layup, dunk
jump, layup, dunk = [], [], []
for i in shot_type.index:
    if 'Jump' in i:
        jump.append(i)
    elif 'Layup' in i:
        layup.append(i)
    elif 'Dunk' in i:
        dunk.append(i)

jump_shots = {'attempted': np.sum(shot_type.loc[jump]['SHOT_ATTEMPTED_FLAG']), 
             'made': np.sum(shot_type.loc[jump]['SHOT_MADE_FLAG']),
             'avg': np.sum(shot_type.loc[jump]['SHOT_MADE_FLAG'])/np.sum(shot_type.loc[jump]['SHOT_ATTEMPTED_FLAG'])}
layup_shots = {'attempted': np.sum(shot_type.loc[layup]['SHOT_ATTEMPTED_FLAG']), 
             'made': np.sum(shot_type.loc[layup]['SHOT_MADE_FLAG']),
             'avg': np.sum(shot_type.loc[layup]['SHOT_MADE_FLAG'])/np.sum(shot_type.loc[layup]['SHOT_ATTEMPTED_FLAG'])}
dunk_shots = {'attempted': np.sum(shot_type.loc[dunk]['SHOT_ATTEMPTED_FLAG']), 
             'made': np.sum(shot_type.loc[dunk]['SHOT_MADE_FLAG']),
             'avg': np.sum(shot_type.loc[dunk]['SHOT_MADE_FLAG'])/np.sum(shot_type.loc[dunk]['SHOT_ATTEMPTED_FLAG'])}
```

```python
df = pd.DataFrame([jump_shots,layup_shots, dunk_shots], index=['jump', 'layup', 'dunk'])
df
```

```python
sns.set_theme(style="darkgrid")
f, ax = plt.subplots(figsize=(5, 5))
sns.set_color_codes("pastel")
sns.barplot(df, x='attempted', y=df.index, label='attempted', color='b')
sns.set_color_codes("muted")
sns.barplot(df, x='made', y=df.index, label='made', color='b')
ax.set_ylabel('shot type')
ax.set_xlabel('number')
plt.show()
```
